##Práctica 1: Distribución normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#Generar datos para la distribución normal
media = 70 #Media
desv_std = 4 #Desviación estándar
num_datos = 500
#Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #mostrar los primeros 50
```

```python
#Crear un rango de valores para crear una densidad
x=np.linspace(media-3 * desv_std, media + 3* desv_std, num_datos)

#Calcular el PDF de la distribución normal
pdf=1 / (desv_std*np.sqrt(2*np.pi))* np.exp(-(x-media)**2/(2*desv_std**2))

#Para no imprimir toda la lista de valores (500), desplegamos los primeros 50
print("Imprimir los primeros 50 valores de x",x[0:50]," y los primeros 50 pdf", pdf[0:50])
```

```python
#Gráfica histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')#Histograma de los datos
plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
#visualizar pdf
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media,2), np.round(desv_std,2)))
plt.title('Funcion de densidad de Probabilidad (PDF)-Distribución normal')
plt.xlabel('Pesos en Kg, de los estudiantes')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Gráficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x,pdf,'k', linewidth=2,)
plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

##Ejercicio 2

```python
plt.figure(figsize=(6,4))
#parametros de la distribución normal
media=70
desv_std=4
total_estudiantes = 500
#variable para cálculo de la probabilidad de que un estudiante que pese menos de 75kg
peso_limite = 79
#primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 utilizando la función de distribución acumulativa (CDF)
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
#cálculo del número de estudiantes que pesan menos de 75kg
estudiantes_menos_75kg = prob_menos_75kg * total_estudiantes
print("Probabilidad de que un estudiante pese menos de 75kg", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75kg:", estudiantes_menos_75kg)
#calcular la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x,media,desv_std)
#Gráficar la distribución normal
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x <= peso_limite), color='skyblue',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$,X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
#parametros de la distribución normal
media=70
desv_std=4
total_estudiantes = 500
#variable para cálculo de la probabilidad de que un estudiante que pese menos de 75kg
peso_limite = 79
#La probabilidad de que un estudiante pese menos o igual a 79kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)
#La probabilidad de que un estudiante pese más de 79kg es complementaria
prob_mas_79kg = 1 - prob_menos_79kg
#cálculo del número de estudiantes que pesan máss de 75kg
estudiantes_mas_79kg = prob_mas_79kg * total_estudiantes
print("Probabilidad de que un estudiante pese mas de 79kg", prob_mas_79kg)
print("Número aproximado de estudiantes que pesan mas de 79kg:", estudiantes_mas_79kg)
#Calcular la funcion de densidad de probabilidad de la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x,media,desv_std)

plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x >= peso_limite), color='skyblue',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 79kg. \n $\mu=70$, $\sigma=4$,X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
