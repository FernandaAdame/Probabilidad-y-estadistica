```python
##Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos.
##Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:
##La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
##Cuantos estudiantes pesan más de 68 y menos de 72 kg.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Los estudiantes pesan menos de 72kg

```python
plt.figure(figsize=(6,4))
media=70
desv_std=4
total_estudiantes = 500
peso_limite = 72
prob_menos_72kg = norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_72kg = prob_menos_72kg * total_estudiantes
print("Probabilidad de que un estudiante pese menos de 72kg", prob_menos_72kg)
print("Número de estudiantes que pesan menos de 72kg:", estudiantes_menos_72kg)
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x,media,desv_std)
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x <= peso_limite), color='skyblue',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 72kg. \n $\mu=70$, $\sigma=4$,X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

Los estudiantes pesan más de 68kg

```python
plt.figure(figsize=(6,4))
media=70
desv_std=4
total_estudiantes = 500
peso_limite = 68
prob_menos_68kg = norm.cdf(peso_limite, media, desv_std)
prob_mas_68kg = 1 - prob_menos_68kg
estudiantes_mas_68kg = prob_mas_68kg * total_estudiantes
print("Probabilidad de que un estudiante pese mas de 68kg", prob_mas_68kg)
print("Número aproximado de estudiantes que pesan mas de 68kg:", estudiantes_mas_68kg)
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x,media,desv_std)
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x >= peso_limite), color='skyblue',alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68kg. \n $\mu=70$, $\sigma=4$,X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

Los estudiantes pesan más de 68kg y menos de 72kg

```python
plt.figure(figsize=(6,4))
media= 70
desviacion_e = 4
estudiantes = 500
limite_mayor = 72
limite_menor = 68
probabilidad_mayor= norm.cdf((limite_mayor - media)/ desviacion_e)
probabilidad_menor= norm.cdf((limite_menor - media)/ desviacion_e)
probabilidadT = probabilidad_mayor - probabilidad_menor
estudiantes_68_72 = probabilidadT*estudiantes
print("Probabilidad de que un estudiante pese mas de 68kg y menos de 72kg", probabilidadT)
print("Número aproximado de estudiantes que pesan mas de 68kg y menos de 72kg:", estudiantes_68_72)
x = np.linspace(media - 4 * desviacion_e, media + 4 * desviacion_e, estudiantes)
y = norm.pdf(x,media,desviacion_e)
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x, y, where=(x >= limite_menor) & (x <= limite_mayor), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68kg y menos de 72kg. \n $\mu=70$, $\sigma=4$,X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
