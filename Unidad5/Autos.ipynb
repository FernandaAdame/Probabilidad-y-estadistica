```python
import pandas as pd
import matplotlib.pyplot as pit
import numpy as np
```

```python
autos2 = pd.read_csv("autos_corregidos.csv")
autos2.head()
```

Regresión Lineal Simple

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
lm = LinearRegression()
#variables independientes o variables predicadores
x= autos2[["city-mpg"]]
#variables dependiente o variable objetivo a producir
y=autos2["price"]
x.head(10)
```

```python
#Entrenar el algoritmo de regresión lineal
lm.fit(x,y)
```

```python
#Realizamos la fase de predicción
Ypredict = lm.predict(x)
```

```python
#Valor de la pendiente
lm.intercept_
```

```python
#Valor del coeficiente o peso de la variable
lm.coef_
```

```python
#Precisión del modelo
lm.score(x,y)
```

Regresión Lineal multiple

```python
from sklearn.model_selection import train_test_split
```

```python
#Dividir el dataset en 2 partes, una par entrenamiento de datos y otra para probar
#Para machine laerning es necesario dividir los datos en 2 partes, 1 parte
#aproximadamente del 70% de los datos, y la otra del 30%
#la primera parte para entrenar el algoritmo de regresión lineal y la otra para probar
#la precision del algoritmo
```

```python
autos2_train, autos2_test = train_test_split(autos2,random_state=0)
```

```python
#ver la correlación entre variables
autos2[["horsepower","engine-size", "city-mpg","wheel-base", "curb-weight", "price"]].corr()
```

```python
#variables predictoras para la fase de entrenamiento
X_autos2_train = autos2_train[["horsepower","engine-size","city-mpg", "wheel-base","bore"]]

#Variable a predecir fase de entrenamiento
y_autos2_train = autos2_train[["price"]]
```

```python
#variables predictoras para la fase de prueba
X_autos2_test = autos2_test[["horsepower","engine-size","city-mpg", "wheel-base","bore"]]

#Variable a predecir fase de prueba
y_autos2_test = autos2_test[["price"]]
```

```python
#Fase de entrenamiento del modelo lineal lm
lm.fit(X_autos2_train, y_autos2_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(X_autos2_train, y_autos2_train)
```

```python
#Fase de prueba del modelo lineal Im usando el método Predict
yhat = lm.predict(X_autos2_test)
```

```python
lm.score(X_autos2_test, y_autos2_test)
```

Evaluar el modelo usando visualización (graficos)

```python
import seaborn as sns
plt.figure(figsize=(6,4))
ax1 = sns.distplot(y_autos2_test["price"], hist=False, color="r")
ax1 = sns.distplot(yhat, hist=False, color="b", ax=ax1)
plt.title("Precio Real vs Precio Predicción")
ax1.legend(labels=["Valor Real", "Valor Predicción"])
plt.show()
plt.close()
```

```python
plt.figure(figsize=(6,4))
ax1= sns.kdeplot(y_autos2_test["price"], multiple="stack", alpha=0.5, color="r")
ax1 = sns.kdeplot(pd.Series(yhat.flatten()), multiple="stack", alpha=0.5,color="r")
plt.title("Precio Real vs Precio Predicción")
ax1.legend(labels=["Valor Real", "Valor Predicción"])
plt.show()
plt.close()
```
