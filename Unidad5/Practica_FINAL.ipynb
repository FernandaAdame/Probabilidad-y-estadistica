##Regresin lineal y correlacin

Objetivo: Determinar si existe una relacin lineal entre el tiempo dedicado al estudio y
el rendimiento acadmico, y cuantificar dicha relacin.



###Grupo 1

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
import seaborn as sns
grupo1 = pd.read_csv("grupo1.csv")
grupo1.head()
```

```python
grupo1.info()
```

```python
grupo1.HorasEstudio.describe()
```

```python
grupo1.Calificacion.describe()
```

```python
plt.hist(grupo1["Calificacion"], bins=10, color="skyblue", edgecolor="black")
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificaciones")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
lm = LinearRegression()
x= grupo1[["HorasEstudio"]]
y= grupo1["Calificacion"]
lm.fit(x,y)
```

```python
Ypredict = lm.predict(x)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
lm.score(x,y)
```

```python
Ypred = 85.310 - 0.1060 * 24
```

```python
Ypred
```

```python
grupo1[["HorasEstudio","Calificacion"]].corr()
```

```python
plt.figure(figsize=(6,4))
sns.set(style="whitegrid")
sns.regplot(x=x, y=y, scatter=True, color='blue')
plt.title('Diagrama de dispersion con linea de regresin')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.show()
```

###Interpretacin de Resultados:

1. Explicar cul es la interpretacin del coeficiente de regresin en trminos de la relacin entre el tiempo de estudio y las calificaciones finales?

Las horas de estudio que realizan los estudiantes no tienen o casi no tienen ninguna relacin, es decir, que no influyen en nada.


2. Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y
direccin de la relacin entre las variables?

Al ser un valor negativo y casi cercano al 0 se puede interpretar como que existe una relacin nula entre las variables, lo cual se confirma al observar el diagrama de dispersin con linea de regresin.

3. El modelo de regresin lineal es estadsticamente significativo?

En este caso no, debido a que nos indica que las horas de estudio no tienen un impacto en las calificaciones, ya que su relacin es casi nula, por lo que no podemos predecir que tipo de resultados obtendremos en el caso del grupo 1.

4. Qu tan confiable es el modelo de regresin para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?

No es muy confiable debido a que el resultado obtenido al evaluar que tan bien se ajusta a los datos arrojo un valor de 33%.
