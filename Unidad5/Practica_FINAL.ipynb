##Regresión lineal y correlación

Objetivo: Determinar si existe una relación lineal entre el tiempo dedicado al estudio y
el rendimiento académico, y cuantificar dicha relación.



###Grupo 1

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
import seaborn as sns
grupo1 = pd.read_csv("grupo1.csv")
grupo1.head()
```

```python
grupo1.info()
```

```python
grupo1.HorasEstudio.describe()
```

```python
grupo1.Calificacion.describe()
```

```python
plt.hist(grupo1["Calificacion"], bins=10, color="skyblue", edgecolor="black")
plt.title("Histograma de Calificaciones")
plt.xlabel("Calificaciones")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
lm = LinearRegression()
x= grupo1[["HorasEstudio"]]
y= grupo1["Calificacion"]
lm.fit(x,y)
```

```python
Ypredict = lm.predict(x)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
lm.score(x,y)
```

```python
Ypred = 85.310 - 0.1060 * 24
```

```python
Ypred
```

```python
grupo1[["HorasEstudio","Calificacion"]].corr()
```

```python
plt.figure(figsize=(6,4))
sns.set(style="whitegrid")
sns.regplot(x=x, y=y, scatter=True, color='blue')
plt.title('Diagrama de dispersion con linea de regresión')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.show()
```

###Interpretación de Resultados:

1. ¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?

Las horas de estudio que realizan los estudiantes no tienen o casi no tienen ninguna relación, es decir, que no influyen en nada.


2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y
dirección de la relación entre las variables?

Al ser un valor negativo y casi cercano al 0 se puede interpretar como que existe una relación nula entre las variables, lo cual se confirma al observar el diagrama de dispersión con linea de regresión.

3. ¿El modelo de regresión lineal es estadísticamente significativo?

En este caso no, debido a que nos indica que las horas de estudio no tienen un impacto en las calificaciones, ya que su relación es casi nula, por lo que no podemos predecir que tipo de resultados obtendremos en el caso del grupo 1.

4. ¿Qué tan confiable es el modelo de regresión para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?

No es muy confiable debido a que el resultado obtenido al evaluar que tan bien se ajusta a los datos arrojo un valor de 33%.
