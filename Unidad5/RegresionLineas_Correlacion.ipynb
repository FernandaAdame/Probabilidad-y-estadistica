```python
from sklearn.linear_model import LinearRegression #Modelo de regresión lineal
import numpy as np

#Datos de entrada
periodos = [2018, 2019, 2020, 2021, 2022, 2023]
inversion = np.array([2,3,5,4,11,5]).reshape(-1,1)
ganancias = [20,25,34,30,40,31]

#Crear y entrenar el modelo de regresion lineal
modelo = LinearRegression()
modelo.fit(inversion,ganancias)

#Predecir la ganancia para el año 2024
ganancia_2024 = modelo.predict([[8]])
print("La ganancia esperada para el año 2024 es:", ganancia_2024[0])

#coeficientes de la regresión
print("Coeficiente de la pendiente (m): ", modelo.coef_[0])
print("Intercepcion de la recta (b): ", modelo.intercept_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Definir tamaño
plt.figure(figsize=(6,4))

#Datos de entrada
inversion = [2,3,5,4,11,5]
ganancias = [20,25,34,30,40,31]

#Crear el diagrama de dispersión con la línea de regresión
sns.set(style="whitegrid")
sns.regplot(x=inversion,y=ganancias,scatter=True, color='blue')
plt.title("Diagrama de Dispersión con Linea de Regresión")
plt.xlabel("Inversion (millones)")
plt.ylabel("Ganafias (millones)")
plt.grid(True)
plt.show()
```

```python
#Datos de entrada
publicidad = np.array([2,3,5,4,11,5]).reshape(-1,1)
ganancias = np.array([20,25,34,30,40,31])

#Datos de entrada
publicidad = np.array([2,3,5,4,11,5]).reshape(-1,1)
ganancias = np.array([20,25,34,30,40,31])

#Crear y entrenar el modelo de regresión lineal
modelo= LinearRegression()
modelo.fit(publicidad,ganancias)

#Predecir la ganancia para una inversión de 8 en publicidad
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)
print("La ganancia esperada para una inversión de 8 en publicidad es:", ganancia_predicha[0])
print("Valor de la intercepción", modelo.intercept_)
print("Valor de la pendiente", modelo.coef_)

#Crear el diagrama de dispersión con la linea de regresión
plt.figure(figsize=(6,5))
plt.scatter(publicidad, ganancias, color='blue', label='Datos')
plt.plot(publicidad, modelo.predict(publicidad), color='red', label='Linea de regresión')
plt.scatter(inversion_prediccion, ganancia_predicha, color='green', marker='x', s=100, label='Predicción(8, {0:.2f})'.format(ganancia_predicha[0]))
plt.title('Pronostico de ganancias a partir de la inversión en publicidad')
plt.xlabel('Inversion en publicidad (millones)')
plt.ylabel('Ganancias Anuales (millones)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr
#Datos de entrada
publicidad = [2,3,5,4,11,5]
ganancias = [20,25,34,30,40,31]
#Calcular el coeficiente de correlación de Pearson
coeficiente_correlacion, p_valor = pearsonr(publicidad,ganancias)
print("Coeficiente de correlación de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```
