```python
from statistics import *
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
import pandas as pd
import math
from collections import Counter #Contar la frecuencia de cada elemento
autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

###Estadísticas descriptivas básicas y medidas de tendencia central


```python
print("Valor mínimo de los precios:", autos.price.min())
```

```python
print("Valor máximo de los precios:", autos.price.max())
```

```python
moda = mode(autos.price)
print ("La moda de los precios es: ",moda)
```

```python
media = autos["price"].mean()
print ("La media de los precios es: ",media)
```

```python
mediana = autos["price"].median()
print ("La mediana de los precios es: ",mediana)
```

```python
desviacion_estandar = autos["price"].std()
print("La desviación estandar de los precios es: ", desviacion_estandar)
```

Estadísticos y cuartiles

```python
autos.price.describe()
```

Histograma de precios de los automóviles basado en la regla de Sturges

```python
n = len(autos["price"])
K = int(1 + math.log2(n))
plt.hist(autos['price'], bins=K, alpha=0.5)
plt.title("Histograma de precios de los automóviles")
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
```

Conclusiones: En este histograma podemos notar fácilmente cómo varían los precios de los automóviles, siendo muy frecuente que cuesten entre 5000 a 10000,y muy raro que suesten entre 25000 a 45000. Esto resalta la importancia de los histogramas, ya que, con solo observar el histograma, podemos interpretar los datos de manera rápida y clara. En cambio, si solo los observamos directamente desde la base de datos, dependiendo de la cantidad de datos que tengamos, resultaría muy difícil y lento poder interpretarlos.

Diagrama de caja y bigotes: Relación estadística entre precio del automóvil y el tipo de vehículo

```python
datos = autos["body-style"]
plt.figure(figsize=(11,7))
sns.boxplot(x=datos, y=autos["price"])
plt.title("Relación entre precio del automóvil y el tipo de vehículo")
plt.xlabel('Tipo de vehículo')
plt.ylabel('Precio')
plt.show()
```

Conclusiones: En este diagrama podemos observar fácilmente cómo varía el precio dependiendo del tipo de vehículo, siendo el hardtop, el convertible y muy raramente el sedan de los vehículos más caros, mientras el hatchback el que puede llegar a ser el más barato. Esto destaca lo útiles que son los diagramas de cajas y bigotes, ya que, si observáramos estos datos directamente en la base de datos, sería muy difícil realizar una comparación. Sin embargo, al haber creado este diagrama en Python, tanto nosotros como cualquier persona que lo vea, puede notar claramente cómo varían los precios.

Diagrama de caja y bigotes: Relación estadística entre la columna del precio del automóvil y tipo tracción ruedas

```python
datosr = autos["drive-wheels"]
plt.figure(figsize=(8,7))
sns.boxplot(x=datosr, y=autos["price"])
plt.title("Relación entre precio del automóvil y el tipo tracción ruedas")
plt.xlabel('Tipo tracción ruedas')
plt.ylabel('Precio')
plt.show()
```

Conclusiones: Al igual que en el anterior diagrama nosotros podemos interpretar muy fácilmente la relación entre el tipo de tracción ruedas y el precio, y como es que el precio tiende a variar gracias a esto. Por ejemplo las rwd suelen ser mucho más caras a comparación de las fwd y 4wd, sin embargo, si observamos las fwd tienden a ser un poco más baratas que las 4wd.

Diagrama de caja y bigotes: Relación estadística entre la columna del precio del automóvil y la ubicación del motor

```python
datosm = autos["engine-location"]
plt.figure(figsize=(8,7))
sns.boxplot(x=datosm, y=autos["price"])
plt.title("Relación entre el precio del automóvil y la ubicación del motor")
plt.xlabel('Ubicación del motor')
plt.ylabel('Precio')
plt.show()
```

Conclusiones: En este diagrama podemos observar cómo los autos que tienen los motores en la parte trasera son mucho más caros en comparación con los que lo tienen en la parte delantera. Además, nos muestra que es muy raro encontrar autos con el motor en la parte delantera que sean caros, pero que si existen algunos casos.
