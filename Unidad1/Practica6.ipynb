```python
from statistics import *
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
import pandas as pd
import math
from collections import Counter #Contar la frecuencia de cada elemento
autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

###Estadsticas descriptivas bsicas y medidas de tendencia central


```python
print("Valor mnimo de los precios:", autos.price.min())
```

```python
print("Valor mximo de los precios:", autos.price.max())
```

```python
moda = mode(autos.price)
print ("La moda de los precios es: ",moda)
```

```python
media = autos["price"].mean()
print ("La media de los precios es: ",media)
```

```python
mediana = autos["price"].median()
print ("La mediana de los precios es: ",mediana)
```

```python
desviacion_estandar = autos["price"].std()
print("La desviacin estandar de los precios es: ", desviacion_estandar)
```

Estadsticos y cuartiles

```python
autos.price.describe()
```

Histograma de precios de los automviles basado en la regla de Sturges

```python
n = len(autos["price"])
K = int(1 + math.log2(n))
plt.hist(autos['price'], bins=K, alpha=0.5)
plt.title("Histograma de precios de los automviles")
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
```

Conclusiones: En este histograma podemos notar fcilmente cmo varan los precios de los automviles, siendo muy frecuente que cuesten entre 5000 a 10000,y muy raro que suesten entre 25000 a 45000. Esto resalta la importancia de los histogramas, ya que, con solo observar el histograma, podemos interpretar los datos de manera rpida y clara. En cambio, si solo los observamos directamente desde la base de datos, dependiendo de la cantidad de datos que tengamos, resultara muy difcil y lento poder interpretarlos.

Diagrama de caja y bigotes: Relacin estadstica entre precio del automvil y el tipo de vehculo

```python
datos = autos["body-style"]
plt.figure(figsize=(11,7))
sns.boxplot(x=datos, y=autos["price"])
plt.title("Relacin entre precio del automvil y el tipo de vehculo")
plt.xlabel('Tipo de vehculo')
plt.ylabel('Precio')
plt.show()
```

Conclusiones: En este diagrama podemos observar fcilmente cmo vara el precio dependiendo del tipo de vehculo, siendo el hardtop, el convertible y muy raramente el sedan de los vehculos ms caros, mientras el hatchback el que puede llegar a ser el ms barato. Esto destaca lo tiles que son los diagramas de cajas y bigotes, ya que, si observramos estos datos directamente en la base de datos, sera muy difcil realizar una comparacin. Sin embargo, al haber creado este diagrama en Python, tanto nosotros como cualquier persona que lo vea, puede notar claramente cmo varan los precios.

Diagrama de caja y bigotes: Relacin estadstica entre la columna del precio del automvil y tipo traccin ruedas

```python
datosr = autos["drive-wheels"]
plt.figure(figsize=(8,7))
sns.boxplot(x=datosr, y=autos["price"])
plt.title("Relacin entre precio del automvil y el tipo traccin ruedas")
plt.xlabel('Tipo traccin ruedas')
plt.ylabel('Precio')
plt.show()
```

Conclusiones: Al igual que en el anterior diagrama nosotros podemos interpretar muy fcilmente la relacin entre el tipo de traccin ruedas y el precio, y como es que el precio tiende a variar gracias a esto. Por ejemplo las rwd suelen ser mucho ms caras a comparacin de las fwd y 4wd, sin embargo, si observamos las fwd tienden a ser un poco ms baratas que las 4wd.

Diagrama de caja y bigotes: Relacin estadstica entre la columna del precio del automvil y la ubicacin del motor

```python
datosm = autos["engine-location"]
plt.figure(figsize=(8,7))
sns.boxplot(x=datosm, y=autos["price"])
plt.title("Relacin entre el precio del automvil y la ubicacin del motor")
plt.xlabel('Ubicacin del motor')
plt.ylabel('Precio')
plt.show()
```

Conclusiones: En este diagrama podemos observar cmo los autos que tienen los motores en la parte trasera son mucho ms caros en comparacin con los que lo tienen en la parte delantera. Adems, nos muestra que es muy raro encontrar autos con el motor en la parte delantera que sean caros, pero que si existen algunos casos.
